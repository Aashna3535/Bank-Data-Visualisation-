```python
#Loading data set
df = pd.read_csv("Banking Data.csv") 
df
```

```python
#Copying data to other dataframe to work with
data = df.copy()
data
```

```python
#Dimension of data
data.shape
```

```python
#Characteristics of data
data.info()
```

```python
#Statistical characteristics of data
data.describe()
```

```python
#Checking number of missing values
data.isnull().sum()
```

```python
#printing records having missing values
data[pd.isnull(data).any(axis=1)]
```

```python
#Removing rows having more than 10 NAN values
data = data[data.isnull().sum(axis=1) < 10]
```

```python
#Printing records having NAN values
data[pd.isnull(data).any(axis=1)]
```

```python
#Replacing NAN values with a string in textual features
pd.options.mode.chained_assignment = None
data["Gender"].fillna('No Gender', inplace = True)
data["Geography"].fillna('No Geography', inplace = True)
```

```python
#Trasnforming text features into Numeric categorical variables
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
data['Geography'] = le.fit_transform(data['Geography'])
data = data.drop(['Surname'], axis = 1)
data = data.drop(['CustomerId'], axis = 1)
data = data.drop(['RowNumber'], axis = 1)
```

```python
data
```

```python
data[pd.isnull(data).any(axis=1)]
```

```python
#Handling missing values by mean, mode values in numerical features
data["Age"].fillna(int(data["Age"].mean()), inplace = True)
data["Balance"].fillna(round(data["Balance"].mean(),2), inplace = True)
data["NumOfProducts"].fillna(data["NumOfProducts"].mode()[0], inplace = True)
data["CreditScore"].fillna(int(data["CreditScore"].mean()), inplace = True)
data["HasCrCard"].fillna(2, inplace = True)
data["EstimatedSalary"].fillna(round(data["EstimatedSalary"].mean(),2), inplace = True)
```

```python
#Number of records having missing value
data.isnull().values.any()
```

```python
#Checking duplicate rows
dupval = data.duplicated()
print(dupval.sum())
```

```python
#Finding most realted features to exited feature
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
```

```python
df_x= data.iloc[:,0:-1]
df_y= data.iloc[:,-1]
```

```python
df_y
```

```python
model.fit(df_x,df_y)
```

```python
print(model.feature_importances_)
```

```python
#Feature importance with respect to Exited feature 
feat_importances = pd.Series(model.feature_importances_, index=df_x.columns)
feat_importances.nlargest(20).plot(kind='barh')
```

```python
#Finding relationships between numeric and 2 categorical data 
# perform groupby
df = data.groupby(['Exited', 'NumOfProducts']).agg(age=("Age", 'median'))
df = df.reset_index()
  
# plot barplot
sns.barplot(x="NumOfProducts",
           y="age",
           hue="Exited",
           data=df)
```

```python
# perform groupby
df = data.groupby(['Exited', 'Geography']).agg(estsalary=("EstimatedSalary", 'mean'))
df = df.reset_index()
  
# plot barplot
sns.barplot(x="Geography",
           y="estsalary",
           hue="Exited",
           data=df)
plt.show()
# perform groupby
df = data.groupby(['Exited', 'Geography']).agg(bal=("Balance", 'mean'))
df = df.reset_index()
  
# plot barplot
sns.barplot(x="Geography",
           y="bal",
           hue="Exited",
           data=df)
plt.show()
# perform groupby
df = data.groupby(['Exited', 'Gender']).agg(bal=("Balance", 'mean'))
df = df.reset_index()
  
# plot barplot
sns.barplot(x="Gender",
           y="bal",
           hue="Exited",
           data=df)
plt.show()
```
